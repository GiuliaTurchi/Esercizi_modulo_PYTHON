* Esaminiamo la differenza tra lo '*StandardCost*' e il '*DealerPrice*' degli ultimi 20 elementi del dataset: c'è un pattern?

* Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50.
  * Per la fascia d'età inferiore a 20, non risulteranno dati, poichè l'età più piccola presente nel dataset parte da 21 anni.

```python
print("Età minima:", diabetes["Age (years)"].min())
print("Età massima:", diabetes["Age (years)"].max())
```

```python
#Definisco fasce età e filtro dati corrispondenti, in un dizionario
fasce_eta = {"<20": diabetes[diabetes["Age (years)"] < 20],                                        #Età inferiore a 20 anni
             "20-30": diabetes[(diabetes["Age (years)"] >= 20) & (diabetes["Age (years)"] < 30)],  #Età tra 20 e 29 anni
             "30-40": diabetes[(diabetes["Age (years)"] >= 30) & (diabetes["Age (years)"] < 40)],  #Età tra 30 e 39 anni
             "40-50": diabetes[(diabetes["Age (years)"] >= 40) & (diabetes["Age (years)"] < 50)],  #Età tra 40 e 49 anni
             ">50": diabetes[diabetes["Age (years)"] > 50]}                                        #Età superiore a 50 anni

print("Selezione dei dati per fasce di età:")        #Stampo intestazione

#Ciclo for per iterare su ogni fascia d'età e visualizzare un'anteprima dati
for fascia, dati in fasce_eta.items():               #`fascia`=chiave del dizionario (es. "<20"), `dati`=subset filtrato di pazienti
    print(f"\nFascia d'età {fascia}:")               #Stampo nome fascia d'età
    display(dati.head())                             #Mostro prime righe dati, per fascia età corrente
```

* Qual è la media della pressione sanguigna diastolica, per le diverse fasce di età? 

```python
#Ciclo for per calcolare e stampare media pressione sanguigna, per ciascuna fascia
print("Media della pressione sanguigna diastolica per fasce di età:", "\n")  #Stampo intestazione generale

for fascia, dati in fasce_eta.items():                                 #Itero su ogni fascia età (chiave) e relativo subset dati (valore)
    avg_diastolic_bp = dati["Diastolic blood pressure (mm Hg)"].mean() #Calcolo media pressione per dati filtrati
    print(f"Fascia {fascia}: {avg_diastolic_bp:.2f}")                  #Stampo risultato, arrotondato a due decimali

#pattern -> crescita graduale della pressione sanguigna con l'avanzare dell'età
```

* Qual è la media della pressione, per ogni singolo anno di età?

```python
#Seleziono anni età uniche
eta_uniche = diabetes["Age (years)"].unique()

#Calcolo media pressione diastolica per ogni anno di età
media_diastolica = {}
for eta in eta_uniche:
    valori_filtrati = diabetes[diabetes["Age (years)"] == eta]["Diastolic blood pressure (mm Hg)"]
    media_diastolica[eta] = round(valori_filtrati.mean(), 2)

#Trasformo risultati in DataFrame, utilizzando .to_frame()
media_diastolica_df = pd.Series(media_diastolica).to_frame(name="Media pressione diastolica")
media_diastolica_df.index.name = "Età"

#Visualizzo DataFrame
display(media_diastolica_df)
```

```python
#Calcolo limiti quartili di 'bmi'
q1, q2, q3, q4 = ins["bmi"].quantile([0.25, 0.5, 0.75, 1.0])
#.quantile -> calcola quantili (dati divisi in intervalli, basati su proporzioni specificate-> 25% (primo quartile), 50% (mediana), ecc)

#Dizionario filtri per ciascun quartile
filtri_quartili = {"Primo quartile": (ins["bmi"] <= q1),
    "Secondo quartile": (ins["bmi"] > q1) & (ins["bmi"] <= q2),
    "Terzo quartile": (ins["bmi"] > q2) & (ins["bmi"] <= q3),
    "Quarto quartile": (ins["bmi"] > q3)}

#Itero sui quartili e calcolo rispettive statistiche, con arrotondamento
for nome_quartile, filtro in filtri_quartili.items():       #.items() -> utilizzato con dizionari per iterare su coppie chiave-valore
    print(f"Statistiche per {nome_quartile}:")              #Stampo intestazione per ogni quartile
    stats = ins.loc[filtro, "charges"].describe().round(2).to_frame()  #Calcolo statistiche, arrotondo e converto in DataFrame
    #.to_frame(): per convertire una Serie in un DataFrame, facilitando la manipolazione o visualizzazione tabellare
    display(stats)                                          #Stampo statistiche come DataFrame
```
